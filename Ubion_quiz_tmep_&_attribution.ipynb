Attrition은 퇴사 여부 데이터를 (yes:1, no:0)으로 데이터를 변환하여 새로운 파생변수에 대입하고 범주별 레코드 수를 계산하시오

```python
df_ex2 = pd.read_csv('/content/drive/MyDrive/유비온 시험 /HR-Employee-Attrition.csv')
df_ex2
```

```python
#Attrition은 퇴사 여부 데이터를 (yes:1, no:0)으로 데이터를 변환
df_ex2['Attrition_0or1'] = df_ex2['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0) # 파생변수에 값을 대입
# apply는 반복 함수의 느낌, 람다는 결과를 바로 출력해 준다. 딕셔너리의 느낌 
```

```python
df_ex2
```

```python
# 새로 만든 파생변수의 각각의 데이터 개수를 세기
counts = df_ex2['Attrition_0or1'].value_counts()
print(counts)
```
