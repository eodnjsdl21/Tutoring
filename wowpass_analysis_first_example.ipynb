```python
# '강의결제' 값이 'X'인 행들의 '관심자격1'과 '관심자격2' 칼럼 값 출력
interests = pd.concat([df.loc[df['강의결제'] == 'X', '관심자격1'], df.loc[df['강의결제'] == 'X', '관심자격2']])
# 중복된 데이터 개수 세기
counts = interests.value_counts()
# 상위 5개 출력
print(counts.head(5))
```
